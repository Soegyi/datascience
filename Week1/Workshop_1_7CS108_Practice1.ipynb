## __Q9.__ Which occupation represents more males than females? ##


```python
ct = data.groupby(['occupation', 'sex']).size().unstack(fill_value=0)
male_more = ct[1] > ct[0]
ct.head()
male_col = 1
female_col = 0
if male_col in ct.columns and female_col in ct.columns:
    male_more = ct[male_col] > ct[female_col]
    result = ct[male_more]
else:
    print('One of the sex categories is missing in current data.')
result
```

```python
data=pd.read_csv('adult.csv');
data=data.sample(n=30000, random_state=44)
data['sex']=data['sex'].astype(str).str.strip(); 
data['occupation']=data['occupation'].astype(str).str.strip(); 
ct=data.groupby(['occupation','sex']).size().unstack(fill_value=0); 
print('sex columns:', ct.columns.tolist()); 
male=ct.get('Male',0); 
female=ct.get('Female',0); 
out=ct.assign(Female=female, Male=male, male_gt_female=male>female)[['Female','Male','male_gt_female']]; 
print(out); 
print('\nOccupations with more males than females:'); 
print(out[out['male_gt_female']].index.tolist())
data.index
```
